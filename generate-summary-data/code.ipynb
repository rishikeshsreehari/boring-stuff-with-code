```python
states = ['Northen Region',
'Southern Region',
'Western Region',
'Eastern Region',
'North-Eastern Region']

for state in states:

    combined_df = pd.DataFrame(index=range(0,8760),columns=['y_17','y_81','y_19','y_20','y_21','y_22'])

    for year,col in zip(years,combined_df.columns):

        sorted_values = pd_data[pd_data.index.year==year][state].sort_values(ascending=False)

        if year == 2020:

            combined_df[col] = sorted_values[~((sorted_values.index.month == 2) & (sorted_values.index.day == 29) & (pd.DatetimeIndex(sorted_values.index).is_leap_year))].values

        else:

            combined_df[col] = sorted_values.values

    normalize_column = lambda col: col / col.max()

    combined_df = combined_df.apply(normalize_column)

    min_variation_ldc = round(combined_df.mean(axis=1).min()*100,1)
    max_variation_ldc = round(combined_df.mean(axis=1).max()*100,1)

    pc90_2022 = round(len(combined_df['y_22'][combined_df['y_22']>=0.9])/8760*100,1)
    pc90_2017 = round(len(combined_df['y_17'][combined_df['y_17']>=0.9])/8760*100,1)
    pc_percent_change = round((pc90_2022-pc90_2017)/pc90_2017*100,1)

    lf_22 = round(combined_df['y_22'].mean()*100,1)
    lf_17 = round(combined_df['y_17'].mean()*100,1)

    if pc_percent_change<0:
        pc_change_bool = 'decrease' 
    else:
        pc_change_bool = 'increase'

    if lf_22<lf_17:
        lf_bool = 'a decrease' 
    else:
        lf_bool = 'an increase'

    if lf_22<lf_17:
        lf_cbool = 'decline' 
    else:
        lf_cbool = 'rise'

    if lf_22<lf_17:
        lf_ccbool = 'higher' 
    else:
        lf_ccbool = 'lower'

#     print(state)
    
#     string_writeup = f'The Load duration curve in the state varies from the maximum range of {max_variation_ldc} % to a minimum \
#     of {min_variation_ldc} %. Further, almost {pc90_2022}% of the overall demand instances have occured more than 90% of the overall duration, which is a {abs(pc_percent_change)}% {pc_change_bool} \
#     over the past 6 years. Overall, the state has witnessed {lf_bool} in the load factor from {lf_17}% to {lf_22}% between 2017 to 2022. \
#     The {lf_cbool} in the load factor is mainly due to the {lf_ccbool} growth rate of peak demand as compared over the average demand.'
    
    string_writeup_v1 = f'The Load duration curve in the region exhibits a maximum variation of {max_variation_ldc} % and a minimum variation \
    of {min_variation_ldc} %. Furthermore, nearly {pc90_2022}% of the total demand instances have occurred for more than 90% of the total duration, resulting in a {abs(pc_percent_change)}% {pc_change_bool} \
    change over the past 6 years. Overall, the state has experienced {lf_bool} in the load factor, ranging from {lf_17}% to {lf_22}% between 2017 and 2022. \
    The {lf_cbool} in the load factor can be primarily attributed to the {lf_ccbool} growth rate of peak demand compared to average demand.'

    string_writeup_v2 = f'The Load duration curve within the region displays a maximum range of {max_variation_ldc} % and a minimum \
    of {min_variation_ldc} % in variation. Additionally, approximately {pc90_2022}% of all demand instances have occurred for over 90% of the overall duration, resulting in a {abs(pc_percent_change)}% {pc_change_bool} \
    change over the past 6 years. Overall, there has been {lf_bool} in the load factor of the state, ranging from {lf_17}% to {lf_22}% between 2017 and 2022. \
    The {lf_cbool} in the load factor is primarily attributed to the {lf_ccbool} growth rate of peak demand compared to average demand.'

    string_writeup_v3 = f'The Load duration curve within the region demonstrates a maximum deviation of {max_variation_ldc} % and a minimum \
    deviation of {min_variation_ldc} %. Moreover, nearly {pc90_2022}% of the total demand instances have occurred for more than 90% of the overall duration, resulting in a {abs(pc_percent_change)}% {pc_change_bool} \
    shift over the past 6 years. Overall, the state has observed {lf_bool} in the load factor, ranging from {lf_17}% to {lf_22}% between 2017 and 2022. \
    The {lf_cbool} in the load factor is predominantly attributed to the {lf_ccbool} growth rate of peak demand compared to average demand.'

    string_writeup_v4 = f'An analysis of the Load duration curve in the region reveals a wide range of variability, with a maximum deviation of {max_variation_ldc}% and a minimum deviation of {min_variation_ldc}%. Notably, approximately {pc90_2022}% of the total demand instances have occurred for more than 90% of the overall duration, resulting in a {abs(pc_percent_change)}% {pc_change_bool} shift over the past 6 years. Examining the load factor, it has demonstrated fluctuations, varying from {lf_17}% to {lf_22}% between 2017 and 2022. These changes in the load factor can be predominantly attributed to the accelerated growth rate of peak demand compared to average demand.'


    selected_format = random.choice([string_writeup_v1, string_writeup_v2,string_writeup_v3,string_writeup_v4])
    
    print(selected_format)
```
